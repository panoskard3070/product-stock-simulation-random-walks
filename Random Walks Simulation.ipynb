```python
#SIMULATION OF A SINGLE RANDOM WALK (PARTICLE MOVES EITHER LEFT OR RIGHT WITH PROBABILITY 50%)

import random
import matplotlib.pyplot as plt

steps=100
initial_position=0
walk=[initial_position]

for i in range(steps):
  if random.randint(0,1)==1:
    step=1
  else:
    step=-1
  initial_position=initial_position+step
  walk.append(initial_position)

print(walk)
plt.plot(walk)
plt.xlabel("Time")
plt.ylabel("Position")
plt.grid(True)
plt.legend(["Walk"])
```

```python
#SIMULATION OF A SINGLE RANDOM WALK (PARTICLE MOVES EITHER LEFT OR RIGHT WITH CUSTOM PROBABILITIES)

import random
import matplotlib.pyplot as plt

x=float(input("Type in the probability to move right:"))
y=float(input("Type in the probability to move left:"))
if x + y > 1:
  print("The sum must be equal to 1.")
else:
  steps=100
  initial_position=0
  walk=[initial_position]

  for i in range(steps):
    if random.choices([-1,1], weights=[y,x])[0]==1:
      step=1
    else:
      step=-1
    initial_position=initial_position+step
    walk.append(initial_position)

  print(walk)
plt.plot(walk)
plt.xlabel("Time")
plt.ylabel("Position")
plt.grid(True)
plt.legend(["Walk"])
```

```python
#SIMULATION OF A SINGLE RANDOM WALK (PARTICLE CAN MOVE LEFT, RIGHT OR STAY AT THE SAME POSITION)

import random
import matplotlib.pyplot as plt

x=float(input("Type in the probability to move right:"))
y=float(input("Type in the probability to move left:"))
z=float(input("Type in the probability to stay at the same position:"))
if x + y + z > 1:
  print("The sum must be equal to 1.")
else:
  steps=100
  initial_position=0
  walk=[initial_position]

  for i in range(steps):
    result = random.choices([-1,0,1], weights=[y,z,x])[0]
    if result==1:
      step=1
    elif result==0:
      step=0
    else:
      step=-1
    initial_position=initial_position+step
    walk.append(initial_position)

  print(walk)
  plt.plot(walk)
  plt.xlabel("Time")
  plt.ylabel("Position")
  plt.grid(True)
  plt.title("Single Random Walk")
```

```python
#SIMULATION OF MULTIPLE RANDOM WALKS (PARTICLE MOVES EITHER LEFT OR RIGHT WITH PROBABILITY 50%)

import random
import matplotlib.pyplot as plt

number_of_walks=5
walks=[]
for j in range(number_of_walks):

  steps=100
  initial_position=0
  walk=[initial_position]

  for i in range(steps):
    if random.randint(0,1)==1:
      step=1
    else:
      step=-1
    initial_position=initial_position+step
    walk.append(initial_position)
  walks.append(walk)

for k in walks:
  print(k)

for walk in walks:
    plt.plot(walk)

plt.title("5 Random Walks")
plt.xlabel("Time")
plt.ylabel("Position")
plt.grid(True)
plt.legend(["Walk 1", "Walk 2", "Walk 3", "Walk 4", "Walk 5"])

```

```python
#SIMULATION OF MULTIPLE RANDOM WALKS (PARTICLE MOVES LEFT, RIGHT OR STAYS IN THE SAME POSITION FOR EACH WALK WITH CUSTOM PROBABILITY + CUSTOM INITIAL POSITION)

import random
import matplotlib.pyplot as plt

x=float(input("Type in the probability to move right:"))
y=float(input("Type in the probability to move left:"))
z=float(input("Type in the probability to stay at the same position:"))
w=int(input("Type in the initial position:"))
if x + y + z > 1:
  print("The sum must be equal to 1.")
else:
  number_of_walks=5
  walks=[]

  for j in range(number_of_walks):

    steps=100
    initial_position=w
    walk=[initial_position]

    for i in range(steps):
      result = random.choices([-1,0,1], weights=[y,z,x])[0]

      if result==1:
        step=1
      elif result==0:
        step=0
      else:
        step=-1

      initial_position=initial_position+step
      walk.append(initial_position)
    walks.append(walk)

  for k in walks:
    print(k)

  for walk in walks:
    plt.plot(walk)

  plt.title("5 Random Walks")
  plt.xlabel("Time")
  plt.ylabel("Position")
  plt.grid(True)
  plt.legend(["Walk 1", "Walk 2", "Walk 3", "Walk 4", "Walk 5"])
```

```python
#SIMULATION OF MULTIPLE RANDOM WALKS (PARTICLE MOVES LEFT, RIGHT OR STAYS IN THE SAME POSITION FOR EACH WALK WITH CUSTOM PROBABILITY + CUSTOM INITIAL POSITION + CUSTOM STEPS FOR EACH WALK)

import random
import matplotlib.pyplot as plt

number_of_walks=int(input("Enter the number of random walks you want to simulate:"))
walks=[]

for j in range(number_of_walks):

  x=float(input("Type in the probability to move right:"))
  y=float(input("Type in the probability to move left:"))
  z=float(input("Type in the probability to stay at the same position:"))
  i=int(input("Type in the initial position:"))
  s=int(input("Type in the number of steps:"))

  if x + y + z > 1:
    print("The sum must be equal to 1.")
  else:
    steps=s
    initial_position=i
    walk=[initial_position]

    for i in range(steps):
      result = random.choices([-1,0,1], weights=[y,z,x])[0]

      if result==1:
        step=1
      elif result==0:
        step=0
      else:
        step=-1

      initial_position=initial_position+step
      walk.append(initial_position)
    walks.append(walk)

for k in walks:
    print(k)

for walk in walks:
    plt.plot(walk)

plt.title(f'{number_of_walks } "Random Walks"')
plt.xlabel("Time")
plt.ylabel("Position")
plt.grid(True)
plt.legend([f"Walk {i+1}" for i in range(number_of_walks)])
```

```python
#PRODUCT STOCK SIMULATION (EACH PRODUCT IS USING ITS OWN RANDOM WALK)

import random
import matplotlib.pyplot as plt

number_of_walks=int(input("Enter the number of random walks you want to simulate:"))
walks=[]

for j in range(number_of_walks):

  x=float(input("Type in the probability to sell a product within an hour/day:"))
  y=float(input("Type in the probability not to sell a product within an hour/day:"))
  i=int(input("Type in the initial stock:"))
  s=int(input("Type in the number of hours/days:"))

  if x + y > 1:
    print("The sum must be equal to 1.")
  else:
    steps=s
    initial_position=i
    walk=[initial_position]

    for i in range(steps):
      result = random.choices([-1,0], weights=[x,y])[0]

      if result==-1:
        step=-1
      else:
        step=0

      initial_position=initial_position+step
      walk.append(initial_position)
      if initial_position==0:
        break
    walks.append(walk)

for k in walks:
    print(k)

for walk in walks:
    plt.plot(walk)

plt.title("Product Stock")
plt.xlabel("Time")
plt.ylabel("Stock")
plt.grid(True)
plt.legend([f"Product {i+1}" for i in range(number_of_walks)])
```
